```python
# Verificar si es seguro colocar una reina en la posición (fila, columna)
def es_seguro(tablero, fila, columna):
    # Verificar la fila
    for i in range(columna):
        if tablero[fila][i] == 1:
            return False
    # Verificar la diagonal superior izquierda
    i, j = fila, columna
    while i >= 0 and j >= 0:
        if tablero[i][j] == 1:
            return False
        i -= 1
        j -= 1
    # Verificar la diagonal inferior izquierda
    i, j = fila, columna
    while i < len(tablero) and j >= 0:
        if tablero[i][j] == 1:
            return False
        i += 1
        j -= 1
    return True

def resolver_util(tablero, columna, soluciones):
    n = len(tablero)
    # Caso base: si todas las reinas están colocadas
    if columna >= n:
        solucion_actual = []
        for fila in tablero:
            fila_copia = fila[:]
            solucion_actual.append(fila_copia)
        soluciones.append(solucion_actual)
        return
    # Recorrer todas las filas en la columna actual
    for fila in range(n):
        if es_seguro(tablero, fila, columna):
            # Colocar una reina en la posición (fila, columna)
            tablero[fila][columna] = 1
            # Resolver recursivamente para la siguiente columna
            resolver_util(tablero, columna + 1, soluciones)
            # Retroceder y probar otra posición
            tablero[fila][columna] = 0

def resolver_n_reinas(n):
    # Crear un tablero vacío
    tablero = []
    for _ in range(n):
        fila = [0] * n
        tablero.append(fila)

    soluciones = []

    # Resolver el problema
    resolver_util(tablero, 0, soluciones)
    print("Soluciones posibles:", len(soluciones))
    # Imprimir todas las soluciones encontradas
    if len(soluciones) > 0:
        for solucion in soluciones:
            for fila in solucion:
                fila_str = ""
                for celda in fila:
                    if celda == 1:
                        fila_str += "R "
                    else:
                        fila_str += "- "
                print(fila_str)
            print()
    else:
        print("No hay solución posible.")

n = 6  # Tamaño del tablero de ajedrez
resolver_n_reinas(n)
```
